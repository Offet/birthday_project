```python
def send_birthday_greetings():
    """
    This function retrieves birthday data from an Excel file, identifies birthdays for the current day,
    and sends email greetings to those individuals. It also updates the 'year_last_wished' column in the
    data file to track when birthday wishes were sent.
    """

    # Load birthday data from Excel spreadsheet
    df = pd.read_excel("bday_dataset.xlsx")

    # Get today's date in a specific format ("%d-%m") for comparison
    today = datetime.datetime.now().strftime("%d-%m")

    # Get the current year (e.g., "2024")
    current_year = datetime.datetime.now().strftime("%Y")

    # Create an empty list to store indices of people who received birthday wishes
    recipients = []
    # Iterate through each person's birthday data
    for index, person in df.iterrows():
        dob = person["dob"]
        # Extract the person's birthday and format it for comparison
        dob_datetime = datetime.datetime.strptime(dob, "%d-%m-%Y")  # Adjust format if needed
        bday_celeb = dob_datetime.strftime("%d-%m")


        # Check if it's indeed the person's birthday and if they haven't received a wish this year
        if (bday_celeb == today) and current_year not in str(person["year_last_wished"]):
            
            # call on the mail function defined earlier
            send_mail(person["email"], "Happy Birthday!", person["message"])
            # Record the person's index for later update
            recipients.append(index)
          
    # Update the 'year_last_wished' column for people who received birthday wishes
    if recipients != None:
        for recipient_index in recipients:
            # Retrieve the previous years when the person received a birthday wish
            prev_years = df.loc[recipient_index, 'year_last_wished']
            prev_years = str(prev_years)
            current_year = str(current_year)
            # Update the 'year_last_wished' column with the previous years and the current year
            df.loc[recipient_index, 'year_last_wished'] = f"{prev_years}, {current_year}"
            

if __name__ == "__main__":
    # This block of code only executes when the script is run directly
    send_birthday_greetings()
```
